Q1. What is the mathematical formula for a linear SVM?

The mathematical formula for a linear Support Vector Machine (SVM) can be represented as follows:

Given a training dataset with input features X and corresponding labels y, where X = [x₁, x₂, ..., xn] and y = [y₁, y₂, ..., yn], and assuming a binary classification problem with class labels -1 and 1, the goal of the linear SVM is to find the optimal hyperplane that separates the two classes in feature space.

The hyperplane is represented by the equation:

w · x - b = 0

where w is the weight vector perpendicular to the hyperplane, x represents a feature vector, · denotes the dot product, and b is the bias term.

For a new instance x, the predicted class label ŷ is determined based on the sign of the discriminant function f(x):

ŷ = sign(f(x))

The discriminant function f(x) is defined as:

f(x) = w · x - b

In linear SVM, the objective is to find the optimal values of w and b that maximize the margin between the hyperplane and the support vectors. The margin is the minimum distance between the hyperplane and the closest data points from each class. This is achieved by solving an optimization problem, typically using quadratic programming or convex optimization techniques.

The optimization problem can be formulated as follows:

Minimize: 1/2 ||w||²

subject to: yᵢ(w · xᵢ - b) ≥ 1 for all training instances (xᵢ, yᵢ)

In this formulation, ||w|| represents the Euclidean norm of the weight vector w, and yᵢ is the class label (-1 or 1) of the training instance xᵢ.

Q2. What is the objective function of a linear SVM?

The objective function of a linear Support Vector Machine (SVM) is to find the optimal hyperplane that maximizes the margin between the hyperplane and the support vectors. This is achieved by minimizing the objective function, which is typically expressed as:

1/2 ||w||² + C * Σ[max(0, 1 - yᵢ(w · xᵢ - b))]

In this objective function:

||w||² represents the squared Euclidean norm of the weight vector w. It controls the magnitude of the weight vector and determines the margin size.

C is the regularization parameter that balances the trade-off between maximizing the margin and minimizing the training errors. It controls the penalty for misclassifying training instances. A higher value of C emphasizes correctly classifying training instances, while a lower value allows for a wider margin at the expense of possible misclassifications.

Σ[max(0, 1 - yᵢ(w · xᵢ - b))] is the sum over all training instances. It captures the training errors or misclassifications made by the SVM. The term max(0, 1 - yᵢ(w · xᵢ - b)) is known as the hinge loss, which penalizes misclassifications and enforces the requirement that the discriminant function f(xᵢ) = w · xᵢ - b has a margin of at least 1 for all correctly classified instances.

The objective function aims to find the values of w and b that simultaneously minimize the norm of the weight vector (to maximize the margin) and minimize the training errors (to ensure correct classification or a margin of at least 1).

Q3. What is the kernel trick in SVM?

The kernel trick in Support Vector Machines (SVM) is a technique that allows SVMs to efficiently handle non-linearly separable data by implicitly mapping the data into a higher-dimensional feature space. It enables SVMs to perform complex non-linear transformations without explicitly computing the transformed feature vectors.

The kernel trick works by introducing a kernel function that replaces the dot product between feature vectors in the original input space with a non-linear mapping function in a higher-dimensional space. Instead of explicitly mapping the data to the higher-dimensional space, the kernel function computes the dot product directly in that space.

Mathematically, given two input feature vectors x and x', the kernel function K(x, x') can be defined as:

K(x, x') = φ(x) · φ(x')

where φ represents the mapping function that maps the input data from the original input space to the higher-dimensional feature space. The kernel function computes the dot product in the feature space without explicitly calculating the transformed feature vectors.

The advantage of using the kernel trick is that it avoids the computational burden and memory requirements of explicitly computing and storing the transformed feature vectors in the higher-dimensional space. Instead, it leverages the kernel function to implicitly compute the dot product between the transformed feature vectors.

Some commonly used kernel functions include:

Linear Kernel: K(x, x') = x · x'
This kernel represents the standard dot product in the original input space and is used for linearly separable data.

Polynomial Kernel: K(x, x') = (γ * (x · x') + r)^d
This kernel maps the data to a polynomial feature space of degree d. The parameters γ, r, and d control the influence of the kernel.

Gaussian (RBF) Kernel: K(x, x') = exp(-γ * ||x - x'||^2)
This kernel maps the data to an infinite-dimensional feature space using the radial basis function (RBF). The parameter γ determines the width of the kernel and influences the smoothness of the decision boundary.

Q4. What is the role of support vectors in SVM Explain with example

In Support Vector Machines (SVM), support vectors play a crucial role in defining the decision boundary and determining the classification of new instances. They are the data points that lie closest to the decision boundary or are misclassified, and they influence the positioning and orientation of the decision boundary.

To understand the role of support vectors, let's consider a simple binary classification example with two classes: positive (+1) and negative (-1). We have a 2D feature space with two features, and the goal is to find a linear decision boundary that separates the two classes.

Suppose we have the following data points in the feature space:

Positive class:
A: (2, 3)
B: (4, 1)

Negative class:
C: (1, 1)
D: (3, 2)

When training an SVM model, the algorithm identifies the support vectors, which are the closest points to the decision boundary or the points that are misclassified.

In this example, the support vectors are points A and C. These points are critical because they define the position and orientation of the decision boundary. The decision boundary is determined by maximizing the margin between the support vectors while minimizing the misclassification errors.

In a linear SVM, the decision boundary is a hyperplane. In this case, the hyperplane is a line that separates the two classes. The support vectors lie on or closest to the decision boundary. In this example, points A and C lie on the decision boundary or are the closest to it.

Once the SVM model is trained, it can classify new instances based on their position relative to the decision boundary. If a new instance falls on one side of the decision boundary, it will be classified as positive, and if it falls on the other side, it will be classified as negative.

Support vectors are crucial because they have the potential to affect the model's generalization ability. SVMs are robust to outliers as they focus on the support vectors, which are representative of the data distribution. The position and properties of the support vectors help ensure a wider margin and reduce the risk of overfitting.

Q6. SVM Implementation through Iris dataset.

~ Load the iris dataset from the scikit-learn library and split it into a training set and a testing setl
~ Train a linear SVM classifier on the training set and predict the labels for the testing setl
~ Compute the accuracy of the model on the testing setl

~ Try different values of the regularisation parameter C and see how it affects the performance of

Bonus task: Implement a linear SVM classifier from scratch using Python and compare its
performance with the scikit-learn implementation.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC

# Load the Iris dataset
iris = datasets.load_iris()

# Split the dataset into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25)

# Train a linear SVM classifier on the training set
clf = LinearSVC(C=1.0)
clf.fit(X_train, y_train)

# Predict the labels for the testing set
y_pred = clf.predict(X_test)

# Calculate the accuracy of the model
accuracy = np.mean(y_pred == y_test)
print("Accuracy:", accuracy)

# Plot the decision boundaries of the trained model
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.plot(X_test[:, 0], clf.decision_function(X_test))
plt.show()

# Try different values of the regularisation parameter C
for C in [0.1, 1.0, 10.0]:
    clf = LinearSVC(C=C)
    clf.fit(X_train, y_train)
    accuracy = np.mean(clf.predict(X_test) == y_test)
    print("C:", C, "Accuracy:", accuracy)
```
